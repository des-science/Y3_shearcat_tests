```python

total_dict=load_obj('everything_you_need_for_response')

%matplotlib inline
import matplotlib.pyplot as plt
f, ax = plt.subplots(1, 2, sharex='all',figsize=(8,3.))#, sharey='row')
pad =0
color1 = 'C0'
color2 = 'C1'
color3 = 'C2'
from matplotlib.ticker import FormatStrFormatter

R_e1 = np.ones(len(total_dict['theta']))*(total_dict['normale_response']['R11_1']*total_dict['normale_response']['R11_2'])
R_e2 = np.ones(len(total_dict['theta']))*(total_dict['normale_response']['R22_1']*total_dict['normale_response']['R22_2'])
        
ax[0].set_title(r'Response $\xi_+$ vs. mean esponse')#,y=1.1)
ff = 10
ax[0].fill_between(total_dict['theta'], ff*R_e1+ff*total_dict['R1']['err'],ff*R_e1-ff*total_dict['R1']['err'], alpha=.2,color = color1,label = r'$<R>^2$ e1')
ax[0].fill_between(total_dict['theta'], ff*R_e2+ff*total_dict['R2']['err'],ff*R_e2-ff*total_dict['R2']['err'], alpha=.2,color = color2,label = r'$<R>^2$ e2')

ax[0].errorbar(total_dict['theta'], ff*total_dict['tt']['xip_R_E1'], ff*total_dict['tt']['cov_xip_R_E1']['err'], label = r'$R{\xi_+}$ e1',color = color1)
ax[0].errorbar(total_dict['theta'], ff*total_dict['tt']['xip_R_E2'], ff*total_dict['tt']['cov_xip_R_E2']['err'], label = r'$R{\xi_+}$ e2',color = color2)



ax[0].set_xlabel(r'$\theta$ (arcmin)')
ax[0].set_ylabel('R (x10)')

miny = min([min(total_dict['tt']['xip_R_E2']),min(total_dict['tt']['xip_R_E1']),min(R_e2-total_dict['R2']['err']),min(R_e1-total_dict['R1']['err'])])
maxy = max([max(total_dict['tt']['xip_R_E2']),max(total_dict['tt']['xip_R_E1']),max(R_e2+total_dict['R2']['err']),max(R_e1+total_dict['R1']['err'])])

dm = ff*(maxy-miny)*0.7
ax[0].set_ylim(ff*miny-dm,dm+ ff*maxy)
plt.xscale('log', nonposx='clip')

ax[1].set_title(r'Response $\xi_-$ (isotropy check)')#,y=1.1)

ax[1].errorbar(total_dict['theta'], ff*total_dict['tt']['xim_R_E1'], ff*total_dict['tt']['cov_xim_R_E1']['err'],label = r'$R{\xi}$ e1',color = color1)
ax[1].errorbar(total_dict['theta'], ff*total_dict['tt']['xim_R_E2'],ff* total_dict['tt']['cov_xim_R_E1']['err'], label = r'$R{\xi}$ e2',color = color2)
ax[1].errorbar(total_dict['theta'], 0.5*(ff*total_dict['tt']['xim_R_E1']+ff*total_dict['tt']['xim_R_E2']), 0.5*ff*(total_dict['tt']['cov_xim_R_E1']['err']+total_dict['tt']['cov_xim_R_E2']['err']), label = r'$R{\xi}$ mean',color = color3)
   
ax[1].set_xlabel(r'$\theta$ (arcmin)')
    
miny = min([min(total_dict['tt']['xim_R_E2']),min(total_dict['tt']['xim_R_E1'])])
maxy = max([max(total_dict['tt']['xim_R_E2']),max(total_dict['tt']['xim_R_E1'])])

    
ax[1].set_ylim(2*ff*miny,2* ff*maxy)
#plt.tight_layout()
plt.xscale('log', nonposx='clip')
ax[1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax[1].tick_params(labelsize=9)
#plt.legend(loc='upper right', bbox_to_anchor=(1.8, 0.8), ncol=1)
ax[1].legend(loc='upper center', bbox_to_anchor=(-0.15, -0.3), ncol=3)
plt.gcf().subplots_adjust(bottom=0.3)
plt.savefig('./figures/xip_tot.pdf',bbox_inches='tight')
plt.show()
```
